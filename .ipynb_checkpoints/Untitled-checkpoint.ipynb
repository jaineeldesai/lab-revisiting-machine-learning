## Model1: Without Balancing the Data and Dropping Any Features

### X-y Split, Train-Test Split

```python
X = pd.concat([numerical,categorical],axis=1)
y= targets['TARGET_B']
```

```python
display(X.head())
display(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

### Splitting Numerical and Normalizing using MinMaxScaler()

```python
X_train_num= X_train.select_dtypes(np.number)
X_test_num=X_test.select_dtypes(np.number)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler= MinMaxScaler().fit(X_train_num)
X_train_norm=pd.DataFrame(scaler.transform(X_train_num),columns=X_train_num.columns)
```

```python
X_test_norm=pd.DataFrame(scaler.transform(X_test_num),columns=X_train_num.columns)
```

### Splitting Categorical and Encoding using One-Hot Encoder

```python
X_train_cat=X_train.select_dtypes(object)
X_test_cat=X_test.select_dtypes(object)

# X_train_cat=X_train_cat.astype('str')
# X_test_cat=X_test_cat.astype('str')
```

```python
X_train_cat.shape
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first').fit(X_train_cat)
train_encoded=encoder.transform(X_train_cat).toarray()
cols = encoder.get_feature_names_out(input_features=X_train_cat.columns)
X_train_encoded= pd.DataFrame(train_encoded,columns=cols)
```

```python
X_test_encoded=encoder.transform(X_test_cat).toarray()
cols = encoder.get_feature_names_out(input_features=X_test_cat.columns)
X_test_encoded= pd.DataFrame(X_test_encoded,columns=cols)
```

### Concatenating the transformed numerical and categorical

```python
X_train_trans=pd.concat([X_train_norm,X_train_encoded],axis=1)
```

```python
X_test_trans=pd.concat([X_test_norm,X_test_encoded],axis=1)
```

### Testing Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score
```

```python
models_list=[LogisticRegression(), KNeighborsClassifier(), DecisionTreeClassifier()]
```

```python
def run_model(models, X_tr, y_tr,X_te, y_te):
    for model in models:
        model.fit(X_tr, y_tr)
        pred_tr=model.predict(X_tr)
        pred_te=model.predict(X_te)
        print(f"{model.__class__.__name__}: Train -> {model.score(X_tr, y_tr)}, Test -> {model.score(X_te, y_te)}")
        print(f"{model.__class__.__name__}: Precision score: Train -> {precision_score(y_tr, pred_tr,zero_division=0)}, Test -> {precision_score(y_te, pred_te,zero_division=0)}")
```

```python
run_model(models_list,X_train_trans,y_train,X_test_trans,y_test)
```

From the above results, we can see that even though the models have high accuracy scores, the precision score is very low due to high imbalance in the data.

## Features Selection

### Variance Threshold

```python
# for now only on normalized train data to treat imbalance
from sklearn.feature_selection import VarianceThreshold 
var_threshold = 0.02  #higher threshold more columns will be dropped
sel = VarianceThreshold(threshold=(var_threshold))

# This drops the columns that have a variance less than this threshold
sel = sel.fit(X_train_norm) #calculates the variance for all the columns
temp = sel.transform(X_train_norm) #removes the columns below the threshold
var_list = list(sel.get_support())
cols=[col[0] for col in zip(X_train_norm.columns, var_list) if col[1] == True]
num_data_train = pd.DataFrame(temp,columns=cols)
print(X_train_norm.shape)
print(num_data_train.shape)
```

```python
323-83 # 240 columns are below the threshold and are dropped
```

```python
#dropping the same columns from x_test_norm
num_data_test=pd.DataFrame(columns=cols)
for col in X_test_norm.columns:
    if col in cols:    
        num_data_test[col]=X_test_norm[col]
```

```python
num_data_test.shape
```

```python
corr_matrix=numerical[cols].corr()
```

```python
corr_matrix.shape
```

```python
#finding columns that are highly correlated
def high_corr(matrix):
    highly_corr=pd.DataFrame()
    
    for col in matrix.columns:
        correlated_cols = matrix[(abs(matrix[col]) >= 0.95)].index   
        correlated_pairs = pd.DataFrame({'Column 1': col, 'Column 2': correlated_cols})  
        highly_corr = pd.concat([highly_corr, correlated_pairs], ignore_index=True)
    highly_corr = highly_corr[highly_corr['Column 1'] < highly_corr['Column 2']]    
    return highly_corr
```

```python
high_corr(corr_matrix)
```

```python
# need to drop 'DW2', 'DW5','DW6','HV2','LFC4'

num_data_train=num_data_train.drop(['DW2', 'DW5','DW6','HV2','LFC4'],axis=1)
```

```python
corr_matrix=num_data_train.corr() #again checking the correlation
high_corr(corr_matrix)

#since we get no columns that means there are no more highly correlated columns
```

```python
num_data_test=num_data_test.drop(['DW2', 'DW5','DW6','HV2','LFC4'],axis=1)
```

```python
X_test_new=pd.concat([num_data_test,X_test_encoded],axis=1)
```

## Imbalance

### In order to resample the data we need to concatenate target and features 

```python
# concatenating numerical, categorical and target all together (only train data)

y_train=y_train.reset_index(drop=True)
trainset=pd.concat([num_data_train,X_train_encoded,y_train],axis=1)
```

```python
trainset.shape
```

```python
trainset['TARGET_B'].value_counts() #there is a high imbalance
```

```python
from sklearn.utils import resample

category_0 = trainset[trainset['TARGET_B'] == 0]
category_1 = trainset[trainset['TARGET_B'] == 1]
```

### Downsampling

```python
category_0_undersampled = resample(category_0,
                                   replace=False,
                                   n_samples = len(category_1))
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
```

```python
data_downsampled['TARGET_B'].value_counts()
```

```python
X_d=data_downsampled.drop('TARGET_B',axis=1)
y_d=data_downsampled['TARGET_B']
```

```python
models_list=[LogisticRegression(), KNeighborsClassifier(), DecisionTreeClassifier()]
```

```python
run_model(models_list,X_d,y_d,X_test_new,y_test)
```

### Upsampling

```python
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
data_upsampled['TARGET_B'].value_counts()
```

```python
X_u=data_upsampled.drop('TARGET_B',axis=1)
y_u=data_upsampled['TARGET_B']
```

```python
run_model(models_list,X_u,y_u,X_test_new,y_test)
```
